```python
from ultralytics import YOLO
import cv2
import numpy as np

def compare_models(img_path):
    # Load models with identical settings
    baseline = YOLO("../models/yolov11/old.pt")
    tuned = YOLO("../models/yolov11/tuned.pt")
    
    # Predict with YOUR working parameters
    base_results = baseline.predict(img_path, imgsz=640, conf=0.5, save=False)
    tuned_results = tuned.predict(img_path, imgsz=640, conf=0.5, save=False)
    
    # Load image once
    img = cv2.imread(img_path)
    h, w = img.shape[:2]
    
    # Draw baseline results (red)
    left_img = img.copy()
    for box in base_results[0].boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        cv2.rectangle(left_img, (x1,y1), (x2,y2), (0,0,255), 2)
        cv2.putText(left_img, 
                   f"{baseline.names[int(box.cls)]} {float(box.conf):.2f}",
                   (x1, max(20,y1-5)), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,255), 2)
    
    # Draw tuned results (green)
    right_img = img.copy()
    for box in tuned_results[0].boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        cv2.rectangle(right_img, (x1,y1), (x2,y2), (0,255,0), 2)
        cv2.putText(right_img,
                   f"{tuned.names[int(box.cls)]} {float(box.conf):.2f}",
                   (x1, max(20,y1-5)),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

    # Combine and save
    comparison = np.hstack((left_img, right_img))
    cv2.imwrite("clean_comparison.jpg", comparison)
    print("Saved clean_comparison.jpg")

    # Add labels
    cv2.putText(comparison, "Baseline", (10, 30), 
               cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    cv2.putText(comparison, "Tuned", (w+10, 30), 
               cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        
    # Display
    cv2.imshow("Model Comparison (Left: Baseline | Right: Tuned)", comparison)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# Usage - same parameters as your working code
compare_models("jauh+takclear.jpg")
```

```python
from ultralytics import YOLO
import torch
import numpy as np

def evaluate_model(model_path, model_name, data_yaml, imgsz=640, device=0):
    """Evaluate a YOLO model and return metrics"""
    model = YOLO(model_path)
    
    metrics = model.val(
        data=data_yaml,
        imgsz=imgsz,
        device=device,
        split='val',
        conf=0.5  
    )
    
    def get_metric(metric):
        if metric is None:
            return 0.0
        if isinstance(metric, (list, tuple, np.ndarray, torch.Tensor)):
            return float(metric[0]) if len(metric) > 0 else 0.0
        return float(metric)
    
    return {
        "Model": model_name,
        "mAP@0.5": get_metric(metrics.box.map50),
        "mAP@0.5:0.95": get_metric(metrics.box.map),
        "Precision": get_metric(metrics.box.p),
        "Recall": get_metric(metrics.box.r),
        "F1": get_metric(metrics.box.f1),
        "Params": sum(p.numel() for p in model.model.parameters()),
        "Inference_speed": float(metrics.speed['inference'])  # ms/img
    }

def print_comparison(results):
    """Print formatted comparison table"""
    print("\n" + "="*80)
    print("📊 Model Comparison: Baseline vs Hyperparameter-Tuned")
    print("="*80)
    print("{:<20} {:<12} {:<12} {:<10} {:<10} {:<10} {:<12} {:<10}".format(
        "Model", "mAP@0.5", "mAP@0.5:0.95", "Precision", "Recall", "F1", "Params (M)", "Speed (ms)"))
    print("-"*80)
    
    for r in results:
        print("{:<20} {:<12.4f} {:<12.4f} {:<10.4f} {:<10.4f} {:<10.4f} {:<12.2f} {:<10.2f}".format(
            r["Model"],
            r["mAP@0.5"],
            r["mAP@0.5:0.95"],
            r["Precision"],
            r["Recall"],
            r["F1"],
            r["Params"]/1e6,
            r["Inference_speed"]
        ))

data_yaml = "data.yaml"
imgsz = 640
device = 0 if torch.cuda.is_available() else "cpu"

results = [
    evaluate_model("old.pt", "Baseline", data_yaml, imgsz, device),
    evaluate_model("tuned.pt", "Tuned", data_yaml, imgsz, device)
]

print_comparison(results)

if len(results) == 2:
    improvements = {
        "mAP@0.5": (results[1]["mAP@0.5"] - results[0]["mAP@0.5"]) / results[0]["mAP@0.5"] * 100,
        "Precision": (results[1]["Precision"] - results[0]["Precision"]) / results[0]["Precision"] * 100,
        "Recall": (results[1]["Recall"] - results[0]["Recall"]) / results[0]["Recall"] * 100,
        "F1": (results[1]["F1"] - results[0]["F1"]) / results[0]["F1"] * 100,
        "Inference_speed": (results[1]["Inference_speed"] - results[0]["Inference_speed"]) / results[0]["Inference_speed"] * 100
    }
    
    print("\n🔍 Hyperparameter Tuning Improvements:")
    for metric, imp in improvements.items():
        print(f"{metric:>12}: {imp:+.2f}%")
```

```python
from ultralytics import YOLO

model = YOLO("../models/yolov11/yolo11n.pt")  

model.train(
    data="../data/data.yaml",
    epochs=150, 
    imgsz=640,
    
    batch=16,  
    lr0=0.001,  
    lrf=0.01, 
    warmup_epochs=5, 
    weight_decay=0.0005,
    momentum=0.9,  
    
    box=0.05,  
    cls=0.5,  
    dfl=1.0,
    
    hsv_h=0.01, 
    hsv_s=0.5,  
    hsv_v=0.3,  
    translate=0.05, 
    scale=0.2, 
    fliplr=0.0,  
    mosaic=0.75, 
    mixup=0.0,  
    copy_paste=0.0, 
    
    patience=35, 
    device=0,  
    overlap_mask=True, 
    mask_ratio=4,  
    close_mosaic=10, 
    
    erasing=0.1, 
    perspective=0.0005,  
    shear=0.0  
)

model.val(
    conf=0.4,  
    iou=0.5,  
    split='val'
)
```
